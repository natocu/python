```python
# 引入 requests 模組
import requests
payload = {"id" : ['1','2','3','4','5']}

# 使用 GET 方式下載普通網頁
r = requests.get('https://jsonplaceholder.typicode.com/users' , params=payload)

print(r.text)
```

```python
# 使用時間的方法
import datetime
nowUtcUnixTime = datetime.datetime.utcnow().timestamp()
print(nowUtcUnixTime)

# 時間格式
nowTime = datetime.datetime.utcnow().strftime("%Y-%m-%d-%H-%M")
print(nowTime)

# 使用當下時區的方法
from pytz import timezone
localTime = datetime.datetime.fromtimestamp(nowUtcUnixTime,timezone('Asia/Taipei'))
print(localTime)

firsttime=localTime.strftime("%Y-%m-%d-%H-%M")
print(firsttime)


#開啟資料夾和寫入方式，注意只能一次性寫入
import os

timefile=open(firsttime,'w')
timefile.write(r.text)
timefile.close()

```
